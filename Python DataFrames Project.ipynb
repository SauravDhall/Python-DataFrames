# Task2: The cost column in the dataframe “Project” has some missing values.compute these missing values. Replace the missing values by running average.

```python
len(b)
```

```python
sum = 0 # Initial sum is zero
count = 0 

# Iteration 
for i in range(len(b)): # length is 14.
    if pd.isna(b.loc[i, 'Cost']): # find is any Nan value in the cost.
        # Replace NaN with running average of previous valid costs
        b.loc[i, 'Cost'] = sum / count
    else:
        # Update sum and count
        sum += b.loc[i, 'Cost']
        count += 1                 # increase the value by 1
# Round 'Cost' to integer value
b['Cost'] = b['Cost'].astype(int) # to get the outcome in interger value.
b
```

# Task3: Split the name column in the Employee dataframe into two new columns “First Name”, and “LastName” and remove the older “name” column.

```python
# split the column 'Name' into two columns i.e. first name and last name by the separator/delimiter 'Space'
A2[['First Name', 'Last Name']] = A2['Name'].str.split(' ', expand=True) 
# drop the column 'Name' from the dataframe
A3 = A2.drop('Name', axis='columns')
A3 # get the desired result 
```

# Task4: Join all three dataframes in one single dataframe. Name it “Final”

```python
#We have two functions for join all dataframes one is join() and other is merge(). for index based, use join() function. 
#But in this case, we have a common column "ID" in all the dataframe so we'll use merge () function for joining all the dataframes.
#As merge() is made for column-based joins.

Final = A3.merge(b, on="ID").merge(c, on="ID") # join all dataframes and assign the name 'Final'
Final # get the result.
```

# Task5: Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects.


```python
# Make a new column 'Bonus' and add 5% to the cost only for those employees who has the status 'Finished'.
Final['Bonus'] = np.where(Final['Status'] == 'Finished', # if the condition met 
                          Final['Cost']*0.05,           # if the consdition is true
                          0)                            # otherwise no bonus
Final['Bonus'] # get the final result
Final
```

 # Task6: Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.

```python
Final_1 = Final.copy()# <-- make a copy because the changes is not updated in the previous table.
Final_1['Designation level'] = np.where(Final_1['Status'] == 'Failed',# the condition if the status is Failed.
                      Final_1['Designation level'] - 1,   # if condition True, designation level reduce by 1.
                      Final_1['Designation level']) # if it's not true, no change in designation level.
Final_1['Designation level']
Final_1 # get the desired output
```

# Task7: Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
"""Two ways to do it one is by using where function"""
# we use the 'where' condition to check the condition, where a condition is True, else replace with something else.
#Alternate way, we can use 'Loc' but 'loc' handled True and False separately. thats the major difference.
Final_2 = Final_1.copy()
Final_2['First Name'] = np.where(Final_2['Gender'] == 'M',# the condition if the Gender is M
                      'Mr. '+ Final_2['First Name'],   # if condition True
                      'Mrs. '+ Final_2['First Name'])# if the condition False
#Final_2['First Name']
Final_2 = Final_2.drop('Gender', axis='columns') # Drop the column Gender
Final_2# get the result 
```

```python
""" Another way by using apply function"""
Final_3 = Final_1.copy()
Final_3['First Name'] = Final_3.apply(
    lambda x: 'Mr. ' + x['First Name'] if (x['Gender'] == 'M' and not x['First Name'].startswith('Mr.'))
    else ('Mrs. ' + x['First Name'] if (x['Gender'] == 'F' and not x['First Name'].startswith('Mrs.'))
          else x['First Name']),
    axis=1
)
Final_3['First Name']
Final_3
```

 # Task8: Promote designation level by 1 for the employees whose age is more than 29 years.


```python
Final_4 = Final_2.copy() # make a copy
def promote(x):          # define a fucntion 'promote'
    if x['Age'] > 29:    # Condition
        return x['Designation level'] + 1   # If the condition is greater than 29 then "+1" in the designation level.
    else:
        return x['Designation level']        # if the condition doesn't meet the condition then no change in the level of designation.

Final_4['Designation level'] = Final_4.apply(promote, axis=1) # apply function to each row 
Final_4 # get the desired output
```
